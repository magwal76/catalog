```python
# @title
import json
from IPython.display import display, HTML
import ipywidgets as widgets

# Load the JSON file (replace with the correct file path if necessary)
with open('it_service_catalog.json', 'r') as file:
    it_service_catalog = json.load(file)

# Extract services from the catalog
services = it_service_catalog['services']

# Get unique categories from services
categories = sorted(set(service['category'] for service in services))

# Output widget to control the output display
output = widgets.Output()

# Function to display selected service details
def display_service_details(service_name):
    selected_service = next(service for service in services if service['name'] == service_name)

    # Clear previous output
    output.clear_output()

    # Extract last update information and verified date
    last_update = selected_service.get('lastupdate', ["", "", ""])
    verified = selected_service.get('verified', '')

    # Get the values
    document_value = selected_service.get('document', '')
    order_value = selected_service.get('order', '')
    form_value = selected_service.get('form', '')

    # Generate HTML based on whether it is a URL or regular text
    document_html = (
        f'<a href="{document_value}" target="_blank" style="font-size: 1.2rem;">Tjänstebeskrivning</a>'
        if document_value.startswith(("http://", "https://"))
        else f'<span style="color: black; font-size: 1.2rem;">{document_value}</span>'
    )

    order_html = (
        f'<a href="{order_value}" target="_blank" style="font-size: 1.2rem;">{order_value}</a>'
        if order_value.startswith(("http://", "https://"))
        else f'<span style="color: black; font-size: 1.2rem;">{order_value}</span>'
    )

    form_html = (
        f'<a href="{form_value}" target="_blank" style="font-size: 1.2rem;">{form_value}</a>'
        if form_value.startswith(("http://", "https://"))
        else f'<span style="color: black; font-size: 1.2rem;">{form_value}</span>'
    )

    # Create HTML formatted service details with CSS for styling
    service_details = f"""
    <head>
        <link rel="preconnect" href="https://fonts.googleapis.com">
        <link rel="preconnect" href="https://fonts.gstatic.com" crossorigin>
        <link href="https://fonts.googleapis.com/css2?family=Roboto:ital,wght@0,100;0,300;0,400;0,500;0,700;0,900;1,100;1,300;1,400;1,500;1,700;1,900&display=swap" rel="stylesheet">
    </head>
    <body>
        <br>
        <p style="color: #003366; text-align: center; font-size: 4rem; font-family: 'Roboto', sans-serif; font-weight: 300; margin: 0;">
            {selected_service['name']}
        </p>
        <div style="background-color: #003366; padding: 20px; border-radius: 10px; position: relative;">
            <div style="display: flex; justify-content: space-between;">
                <div style="background-color: white; padding: 15px; margin: 10px 0; border-radius: 8px; width: 25%;">
                    <p style="color: #003366; font-size: 1.9rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                        Tjänsteområde
                        <br>
                        <span style="color: black; font-size: 1.2rem;">{selected_service['category']}</span>
                    </p>
                </div>
                <div style="background-color: #003366; padding: 0px; display: flex; flex-direction: column; justify-content: space-between;">
                    <div>
                        <h2 style="color: white; text-align: right; font-size: 28px; margin: 0;">
                            <span style="font-size: 14px; font-style: italic; font-family: 'Roboto', sans-serif; font-weight: 100;">
                                Id: {selected_service['id']}
                            </span>
                        </h2>
                    </div>
                    <div>
                        <p style="color: white; font-size: 14px; font-style: italic; text-align: right; font-family: 'Roboto', sans-serif; font-weight: 100; margin-top: 5px;">
                            Senast uppdaterad:
                            <span>{last_update[0]} av {last_update[1]} (Id: {last_update[2]})</span>
                        </p>
                    </div>
                </div>
            </div>
            <div style="background-color: white; padding: 15px; margin: 10px 0; border-radius: 8px;">
                <p style="color: #003366; font-size: 1.9rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                    Kort beskrivning
                    <br>
                    <span style="color: black; font-size: 1.2rem;">{selected_service['description']}</span>
                </p>
            </div>
            <div style="background-color: white; padding: 15px; margin: 10px 0; border-radius: 8px;">
                <p style="color: #003366; font-size: 1.9rem; font-family: 'Roboto', sans-serif; font-weight: 300">Tjänstebeskrivning</p>
                <br>
                <p style="color: #003366; font-size: 1.5rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                    Relaterade dokument
                    <br>
                    {document_html}
                </p>
                <br>
                <p style="color: #003366; font-size: 1.5rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                    Leverantör
                    <br>
                    <span style="color: black; font-size: 1.2rem;">{selected_service['provider']}</span>
                </p>
                <br>
                <p style="color: #003366; font-size: 1.5rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                    Prismodell
                    <br>
                    <span style="color: black; font-size: 1.2rem;">{selected_service['cost']}</span>
                </p>
                <br>
                <p style="color: #003366; font-size: 1.5rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                    SLA
                    <br>
                    <span style="color: black; font-size: 1.2rem;">{selected_service['sla']}</span>
                </p>
                <br>
                <p style="color: #003366; font-size: 1.5rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                    Beställningskanal
                    <br>
                    {order_html}
                </p>
                <br>
                <p style="color: #003366; font-size: 1.5rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                    Beställningsformulär
                    <br>
                    {form_html}
                </p>
            </div>
            <div style="display: grid; grid-template-columns: 1fr 1fr; gap: 30px; margin-top: 30px; padding: 0px;">
                <div style="background-color: white; padding: 10px; border-radius: 8px;">
                    <p style="color: #003366; font-size: 1.9rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                        Tjänsteägare leverantör
                        <br>
                        <span style="color: black; font-size: 1.2rem;">{selected_service['ownervendor']}</span>
                    </p>
                </div>
                <div style="background-color: white; padding: 10px; border-radius: 8px;">
                    <p style="color: #003366; font-size: 1.9rem; font-family: 'Roboto', sans-serif; font-weight: 300">
                        Tjänsteägare Trafikförvaltningen
                        <br>
                        <span style="color: black; font-size: 1.2rem;">{selected_service['ownercustomer']}</span>
                    </p>
                </div>
            </div>
        </div>
    </body>
    """

    # Display the new service details in the output widget
    with output:
        display(HTML(service_details))

# Create a dropdown widget for categories
category_dropdown = widgets.Dropdown(
    options=categories,
    description='Tjänsteområde',
    layout=widgets.Layout(width='max-content')
)

# Create a dropdown widget for service names (empty for now)
service_dropdown = widgets.Dropdown(
    options=[],
    description='Tjänst',
    layout=widgets.Layout(width='max-content')
)

# Apply a custom style to ensure both the label (description) and widget are left-aligned
category_dropdown.style.description_width = '100px'
service_dropdown.style.description_width = '100px'

# Function to update services based on selected category
def update_service_dropdown(category):
    # Get the services in the selected category
    filtered_services = [service['name'] for service in services if service['category'] == category]
    service_dropdown.options = filtered_services

    # If there are services available, select the first one
    if filtered_services:
        service_dropdown.value = filtered_services[0]
        display_service_details(filtered_services[0])

# Update service dropdown when a category is selected
def on_category_change(change):
    update_service_dropdown(change['new'])

# Display the selected service when the service dropdown is changed
def on_service_change(change):
    display_service_details(change['new'])

# Observe changes in the category dropdown
category_dropdown.observe(on_category_change, names='value')

# Observe changes in the service dropdown
service_dropdown.observe(on_service_change, names='value')

# Display the dropdowns and output
display(category_dropdown)
display(service_dropdown)
display(output)

# Initialize with the first category
if categories:
    category_dropdown.value = categories[0]
    update_service_dropdown(categories[0])
```
